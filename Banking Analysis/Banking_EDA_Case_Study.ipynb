```python
# Defining Income Band Boudaries
bins = [0, 100000, 300000, float('inf')]
labels = ['Low Income', 'Medium Income', 'High Income']

# Creating Income Band column
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels)
```

```python
# Examining the distribution of unique categories in categorical columns

categorical_cols = ['Banking Contact',
                    'Nationality',
                    'Occupation',
                    'Fee Structure',
                    'Loyalty Classification',
                    'Properties Owned',
                    'Risk Weighting',
                    'BRId',
                    'GenderId',
                    'IAId',
                    'Income Band']

for col in categorical_cols:
  print(f"\nValue Counts for '{col}':")
  display(df[col].value_counts())
```

```python
missing_values = df.isnull().sum()
print("Missing Values in Each Column:\n", missing_values)
```

```python
df['Joined Bank'] = pd.to_datetime(df['Joined Bank'], format='%d-%m-%Y')
print(df['Joined Bank'].dtype)
```

```python
# Numeriacal Analysis

numerical_cols = ['Age',
                  'Estimated Income',
                  'Superannuation Savings',
                  'Credit Card Balance',
                  'Bank Loans',
                  'Bank Deposits',
                  'Checking Accounts',
                  'Saving Accounts',
                  'Foreign Currency Account',
                  'Business Lending']
```

```python
# Univariate Analysis and Visualization

plt.figure(figsize=(15, 20))
for i, col in enumerate(numerical_cols):
    plt.subplot(5, 2, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

```

```python
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Biivariate analysis and visualization

pairs_to_plot = [
    ('Age', 'Estimated Income',),
    ('Age', 'Superannuation Savings'),
    ('Bank Deposits', 'Saving Accounts',),
    ('Checking Accounts', 'Saving Accounts',),
    ('Checking Accounts', 'Foreign Currency Account',),
    ('Estimated Income', 'Checking Accounts',),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans',)
]

for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(8, 6))
    sns.regplot(x=x_col, y=y_col, data=df, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
    plt.title(f'Relationship Between {x_col} vs {y_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()


```

# KEY INSIGHTS

## Strongest Positive Correlations:
### Bank Deposits & Checking Accounts: 0.84
- People with higher bank deposits tend to maintain higher balances in their
  checking accounts.

###Checking Accounts & Saving Accounts: 0.75

- A strong relationship, indicating customers who use checking accounts also likely have savings accounts.

### Bank Deposits & Saving Accounts: 0.75

- Logical, as people with higher deposits are also likely to keep funds in savings.

### Bank Loans & Business Lending: 0.42

- Individuals or businesses with general bank loans are also involved in business lending.

## Moderate Correlations:
- Credit Card Balance & Bank Deposits: 0.38

- Credit Card Balance & Bank Loans: 0.37

- Foreign Currency Account & Business Lending: 0.37

These moderate values suggest some behavioral or financial overlaps but not strong dependencies.

## Interesting Relationships:
### Estimated Income has moderate correlation with:

- Superannuation Savings (0.37)

- Credit Card Balance (0.30)

- Bank Loans (0.33)

- Business Lending (0.33)

Suggests that higher income is loosely linked to broader financial product engagement.

### Age is almost uncorrelated with all financial features.

- Indicates financial behavior in this dataset is not strongly age-dependent.

## Weak/No Correlations:
- Values close to 0 like between Age & most variables, and Superannuation Savings & Bank Loans (0.24) indicate no strong linear relationship.

## Business Implications:
- Bundling Products: Consider bundling checking, savings, and deposit accounts in marketing—they often go hand-in-hand.

- Credit & Loans: Customers with higher credit card balances could be targeted for personal loan offers.

- Cross-sell Opportunities: Business lending and bank loans can be cross-promoted together.

- Personalization: Since age doesn’t correlate much, targeting should be based on income and behavior, not age.
